```python
# Load the wikipedia dataset

ds = load_dataset("wikimedia/wikipedia","20231101.en")
```

```python
# Verify dataset structure
ds
```

```python
ds_train = ds['train']
ds_train[0]['title']
```

```python
# Select filtration terms to filter the pages on wikipedia

data_filtration_terms = [
    # Core filtration terms
    "filtering", "smoothing", "denoising", "outlier", "thresholding",
    "calibration", "normalization", "compression", "decimation", "downsampling",
    "rejection", "validation", "cleaning", "processing", "aggregation",

    # Sensors & perception
    "lidar", "radar", "sonar", "ultrasonic", "imu", "odometry", "camera",
    "thermal", "depth", "sensing", "sensors", "perception", "vision",
    "imaging", "pointcloud", "reflection", "echo", "noise", "artifact",

    # Algorithms & methods
    "kalman", "particle", "bayesian", "median", "lowpass", "highpass",
    "voxelization", "segmentation", "clustering", "extraction", "feature",
    "ransac", "hdr", "dsp", "transform", "convolution", "regression",

    # Robotics fundamentals
    "robot", "autonomy", "navigation", "localization", "mapping", "slam",
    "path", "planning", "avoidance", "traction", "control", "dynamics",
    "kinematics", "actuator", "motor", "encoder", "battery", "payload",
    "autonomous", "vehicles", "agv", "amr",

    # Industrial context
    "facility", "factory", "warehouse", "logistics", "assembly", "line",
    "material", "handling", "transport", "pallet", "forklift", "conveyor",
    "inventory", "storage", "retrieval", "crossdock", "manufacturing",

    # System & operations
    "fleet", "management", "coordination", "scheduling", "orchestration",
    "monitoring", "diagnostics", "maintenance", "telemetry", "update",

    # Technical domains
    "programming", "software", "algorithm", "firmware", "middleware",
    "hardware", "electronics", "circuit", "network", "protocol", "edge",
    "computing", "cloud", "iot", "cybersecurity", "latency", "bandwidth",

    # Performance
    "efficiency", "optimization", "accuracy", "precision", "reliability",
    "robustness", "scalability", "throughput", "performance", "resource",

    # Safety & standards
    "safety", "standards", "certification", "compliance", "redundancy",
    "failover", "risk", "hazard", "zones", "collision", "emergency",

    # Development tools
    "simulation", "digital", "twin", "modeling", "debugging", "profiling",
    "testing", "validation", "deployment", "integration", "ros", "api",

    # Industry sectors
    "automotive", "aerospace", "pharmaceutical", "ecommerce", "fulfillment"
]


def filter_func_count(input_item):
  # Will filter to only return pages that have atleast 5 hits
  # Change this to extact desired size of dataset
    if input_item['hits'] > 4:
        return True
    return False

def filter_mod(input_item):
    textset = set(input_item['text'].lower().split())
    counter = 0
    for item in data_filtration_terms:
        if item in textset:
            counter += 1
    input_item['hits'] = counter
    return input_item


```

```python
dsf = ds_train.map(filter_mod,batched=False,remove_columns = ['id','url','title'] )
```

```python
# Find number of pages

hit_count = dict()
for item in dsf:
  hit_number = item['hits']
  hit_count[hit_number] = hit_count.get(hit_number,0) + 1
```

```python
# Sort based on quantity of hits, see number of pages for each hitcount value
test_list = [(key,item) for key,item in hit_count.items()]
test_list.sort(key = lambda x : x[1],reverse = True)
test_list[:25]
```

```python
dsfocused = dsf.filter(filter_func_count)
```
